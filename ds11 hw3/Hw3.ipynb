```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Housing.csv')
df.head(10)
```

```python
#напишіть функцію гіпотези лінійної регресії у векторному вигляді
def hipotes_func(X,w):
  return np.dot(X,w)
```

```python
#створіть функцію для обчислення функції втрат у векторному вигляді
def loss_func(w,X,y):
  length = len(y)
  predicted = hipotes_func(X,w)
  squared_errors = (predicted - y) ** 2
  mse = np.sum(squared_errors) / (2 * length)
  return mse
```

```python
#реалізуйте один крок градієнтного спуску
def step_gradient(X, y, w, learning_rate):

    length =len(y)
    predicted = hipotes_func(X, w)
    error = predicted - y
    gradient = np.dot(X.T, error) / length
    w = w - learning_rate * gradient
    return w
```

```python
def gradient_descent(X, y, w, learning_rate, num_iter, eps):

    ones = np.ones((X.shape[0], 1))
    X = np.hstack((ones, X))

    loss = loss_func(X, y, w)
    loss_list = [loss]

    for _ in range(num_iter):
        w = step_gradient(X, y, w, learning_rate)

        loss = loss_func(X, y, w)
        if abs(loss - loss_list[-1]) < eps:
            loss_list.append(loss)
            break

        loss_list.append(loss)

    return w, loss_list
```

```python
def normalize(data):
    result = []
    for x in data:
        norm_x = (x - np.mean(data))/(np.max(data) - np.min(data))
        result.append(norm_x)
    return result

norm_data = pd.DataFrame()
columns = ['price', 'area', 'bedrooms', 'bathrooms']

for col in columns:
    norm_data[col] = normalize(df[col])
```

```python
#знайдіть найкращі параметри для датасету прогнозуючу ціну на будинок залежно від площі, кількості ванних кімнат та кількості спалень
X = norm_data[['area', 'bedrooms', 'bathrooms']].values
y = norm_data['price'].values
w = np.array([0.0, 0.0, 0.0, 0.0])
learning_rate = 0.001
num_iter = 1000000
eps = 0.000000000001
w, loss_list = gradient_descent(X, y, w, learning_rate, num_iter, eps)
print('Parameters:', w)
```

```python
#знайдіть ці ж параметри за допомогою аналітичного рішення
# Створення моделі лінійної регресії
model = LinearRegression()
```

```python
# Навчання моделі на даних
model.fit(X, y)
```

```python
# Вивід оптимальних параметрів
print("Коефіцієнти вагів w:", model.coef_)
print("Значення вільного члена w0:", model.intercept_)
```

```python
#порівняйте отримані результати
print(f"різниця між показниками вільного члена складає {((model.intercept_-w[0])/model.intercept_)*100} %")
w1_different = ((model.coef_[0]-w[1])/model.coef_[0])*100
print(f"різниця між показниками w1 ваг складає {w1_different} %")
w2_different = ((model.coef_[1]-w[2])/model.coef_[1])*100
print(f"різниця між показниками w2 ваг складає {w2_different} %")
w3_different = ((model.coef_[2]-w[3])/model.coef_[2])*100
print(f"різниця між показниками w3 ваг складає {w3_different} %")
```
